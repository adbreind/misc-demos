Enumerate 2019 Yellow Cab files

```python
import s3fs
fs = s3fs.S3FileSystem(anon=True)
trips = fs.ls('nyc-tlc/trip data', detail=True)
files = ['s3://' + trip['Key'] for trip in trips if ('yellow_tripdata_2019' in trip['Key'])]
files
```

Grab just the first 3

```python
import pandas as pd
```

```python
%%time

ldf = pd.concat([pd.read_csv(f) for f in files[:3]], ignore_index = True)
```

Get all stat means grouped by pax count

```python
%%time

ldf.groupby("passenger_count").mean()
```
